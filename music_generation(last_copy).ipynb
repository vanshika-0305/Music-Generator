```python
import zipfile
import os

# Path to your ZIP file
zip_file_path = '/content/maestro-v3.0.0-midi.zip'

# Destination folder to extract the contents
extract_path = '/content/maestro-v3.0.0/'

# Extract the contents
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# Verify extraction
print(f"Contents of {extract_path}:")
print(os.listdir(extract_path))
```

```python
midi_files = []
for root, dirs, files in os.walk(extract_path):
    for file in files:
        if file.endswith(".midi") or file.endswith(".mid"):
            midi_files.append(os.path.join(root, file))

print(f"Total MIDI files found: {len(midi_files)}")
```

```python
import random
import pretty_midi

def extract_notes(midi_file):
    try:
        print(f"Processing: {midi_file}")
        midi_data = pretty_midi.PrettyMIDI(midi_file)
        notes = []
        for instrument in midi_data.instruments:
            if not instrument.is_drum:  # Exclude drums
                notes.extend([note.pitch for note in instrument.notes])
        return notes
    except Exception as e:
        print(f"Error processing {midi_file}: {e}")
        return []

# Randomly sample 5 files from the list to ensure diversity
max_files = 5
limited_midi_files = random.sample(midi_files, max_files)

# Extract notes from the limited number of MIDI files
all_notes = []
for file in limited_midi_files:
    all_notes.extend(extract_notes(file))

# Print the total number of notes extracted
print(f"Total notes extracted from {max_files} files: {len(all_notes)}")
```

```python


# Proceed with preparing sequences, training, etc.
sequence_length = 50  # You can adjust this as needed
vocab = sorted(set(all_notes))  # Unique notes
note_to_int = {note: num for num, note in enumerate(vocab)}  # Map notes to integers

# Prepare sequences
input_sequences = []
output_notes = []
for i in range(len(all_notes) - sequence_length):
    seq_in = all_notes[i:i + sequence_length]
    seq_out = all_notes[i + sequence_length]
    input_sequences.append([note_to_int[note] for note in seq_in])
    output_notes.append(note_to_int[seq_out])

# Normalize input data
X = np.reshape(input_sequences, (len(input_sequences), sequence_length, 1)) / float(len(vocab))
y = tf.keras.utils.to_categorical(output_notes, num_classes=len(vocab))

from tensorflow.keras.callbacks import EarlyStopping

# Define a simplified model
model = Sequential([
    LSTM(128, input_shape=(X.shape[1], X.shape[2]), return_sequences=True),
    Dropout(0.3),
    LSTM(128, return_sequences=True),
    Dropout(0.3),
    LSTM(128),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(len(vocab), activation='softmax')
])

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam')

# Early stopping to avoid overtraining
early_stopping = EarlyStopping(monitor='loss', patience=3, restore_best_weights=True)

# Summarize the model
model.summary()

# Train the model
history = model.fit(
    X, y,
    batch_size=64,  # Reduced batch size
    epochs=1,  # Reduced epochs for quicker training
    callbacks=[early_stopping],
    verbose=1
)

weights_file = "music_model_weights.weights.h5"
model.save_weights(weights_file)
print(f"Weights saved to: {weights_file}")
```

```python
def generate_music(model, int_to_note, start_sequence, num_notes=100):
    generated = []
    pattern = start_sequence

    for _ in range(num_notes):
        prediction_input = np.reshape(pattern, (1, len(pattern), 1)) / float(len(vocab))
        prediction = model.predict(prediction_input, verbose=0)[0]
        index = np.argmax(prediction)
        result = int_to_note[index]
        generated.append(result)

        pattern.append(index)
        pattern = pattern[1:]

    return generated

int_to_note = {num: note for note, num in note_to_int.items()}
start_seq = input_sequences[np.random.randint(len(input_sequences))]
generated_notes = generate_music(model, int_to_note, start_seq)

print("Generated Notes:", generated_notes)
```

```python
def notes_to_midi(notes, output_file="output.mid"):
    midi = pretty_midi.PrettyMIDI()
    instrument = pretty_midi.Instrument(program=0)  # Piano

    start = 0
    for note in notes:
        note_number = int(note)
        note_obj = pretty_midi.Note(velocity=100, pitch=note_number, start=start, end=start + 0.5)
        instrument.notes.append(note_obj)
        start += 0.5

    midi.instruments.append(instrument)
    midi.write(output_file)

notes_to_midi(generated_notes)
```
